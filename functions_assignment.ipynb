### Task 1

```python
def apply_discount(price,discount_percent=5):
    if discount_percent>60:
        return "discount percentage can't be exceeded 60%"
    return price-price*(discount_percent/100)
print(apply_discount(1000,10))
print(apply_discount(500))
apply_discount(900,80)
```

## Task 2

```python
def factorial(n):
    if(n==0 or n == 1):
        return 1
    if(n<0):
        return "it is not possible to calculate the factorial of a negative number"
    return n*factorial(n-1)
print(factorial(5))
print(factorial(0))
factorial(-3)
```

```python
def process_prices(prices):
    discounted_prices = list(map(lambda prices:prices*0.9,prices))
    print(f"discounted_prices : {discounted_prices}")
    filtered_prices =  list(filter(lambda discounted_prices:discounted_prices>300,discounted_prices))
    print(f"filtered_prices : {filtered_prices}")

process_prices([100,500,900,50,750])
```

### Task 7

```python
def add_price(prices_list,price):
    prices_list.append(price)

def get_average_price(prices_list):
    if len(prices_list)==0:
        return 0
    total = 0
    for price in prices_list:
        total = total + price
    return total/len(prices_list)

def get_max_price(prices_list):
    if len(prices_list)==0:
        return 0
    max_price = prices_list[0]
    for price in prices_list:
        if price>max_price:
            max_price=price
    return max_price
```

```python
prices = []
while True:
    print("\nMENU\n")
    print("1->Add price")
    print("2->Show average price")
    print("3->Show highest price")
    print("q->Quit")

    choice = input("Enter your choice: ")

    if choice == "1":
        price = float(input("Enter the price: "))
        add_price(prices,price)

    elif choice == "2":
        avg = get_average_price(prices)
        print(f"Average price: {avg}")

    elif choice == "3":
        highest = get_max_price(prices)
        print(f"Highest price: {highest}")

    elif choice == "q":
        print("Exit")
        break

    else:
        print("Invalid choice")
```
